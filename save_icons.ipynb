```python
import pandas as pd
import urllib.request
from datetime import datetime as dt
```

```python
df = pd.read_csv('full_list_v2.csv')
```

```python
df.head()
```

```python
now = dt.now()

for i in range(0, df.shape[0]):#df.shape[0]):
    try:
        urllib.request.urlretrieve(df.icon_URL[i], 
                    f"src/icons/{(df.title[i]).replace(' ', '').replace('(', '').replace(')', '').lower()}.jpg")
    except:
        pass
    
    if i % 10 == 0:
        print(f'iteration {i} time {dt.now() - now}')
```
